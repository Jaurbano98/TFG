# Codigo de la figura 2.7: Representacion de funciones Delta en distintos instantes de tiempo



# 0. Importar bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1. Creación de funciones

## delta_function:
Función que representa una función delta gaussiana centrada en `a` con un ancho de delta dado.

### Parámetros:
- `x`: Variable independiente.
- `a=0`: Centro de la función delta.
- `delta_width=0.1`: Ancho de la función delta.

### Retorna:
- Valor de la función delta gaussiana en el punto `x`.

```python
def delta_function(x, a=0, delta_width=0.1):
    return np.exp(-((x - a) / delta_width)**2) / (delta_width * np.sqrt(np.pi))
```

# scale_to_max:

Función para normalizar el área.

```python
def scale_to_max(y, target_max):
    max_value = np.max(y)
    scale_factor = target_max / max_value
    return y * scale_factor
```

# 2. Variables de entorno

- `x`: Crear un espacio de variable independiente de -5 a 5 con 1000 puntos.
- `y`: Aplicar la función `delta_function` a cada punto de `x`, almacenando los resultados en `y`.
- `t1 = 1`: Punto de tiempo 1 para la función delta.
- `t2 = 3`: Punto de tiempo 2 para la función delta.
- `y_t1`: Aplicar la función delta con el centro en `t1` a cada punto de `x`, almacenando los resultados en `y_t1`.
- `y_t2`: Aplicar la función delta con el centro en `t2` a cada punto de `x`, almacenando los resultados en `y_t2`



```python
x = np.linspace(-5, 5, 1000)
y = delta_function(x)
y_scaled = scale_to_max(y, target_max=1)
t1 = 1
t2 = 3
y_t1 = delta_function(x, a=t1)
y_t1_scaled = scale_to_max(y_t1, target_max=1)
y_t2 = delta_function(x, a=t2)
y_t2_scaled = scale_to_max(y_t2, target_max=1)
```

# 3. Dibujo de la funcion delta de Dirac en los instantes t0, t1, y t2

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(x, y_scaled, label='Función $\delta(t = 0)$')
ax.plot(x, y_t1_scaled, label=r'Función $\delta(t - t1) $, t1 = 1', linestyle='dashed')
ax.plot(x, y_t2_scaled, label=r'Función $\delta(t - t1) $, t2 = 3', linestyle='dashed')
ax.set_xlabel('t')
ax.set_ylabel(r'$\delta$')
ax.legend()
plt.text(-0.1, 0.1, '$t_0$', color='r')
plt.text(0.9, 0.1, '$t_1$', color='r')
plt.text(2.9, 0.1, '$t_2$', color='r')
plt.show()
```
