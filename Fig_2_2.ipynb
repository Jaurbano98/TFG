# 0. Importar bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1. Creación de funciones

```python
def dv_dt(v, t):
    return -v
```

# 2. Variables de entorno

## 2.1. Solución analitica
t = Crear un espacio de tiempo de 0 a 10 con 100 puntos \
v(0) = 1 es el valor inicial del potencial \



```python
t = np.linspace(0, 10, 70)
v0 = 1.0
v_analitica = v0 * np.exp(-t)
```

## 2.2. Solución analitica

Inicialización del primer valor de v_numerica con v0.

Bucle para calcular los valores siguientes de v_numerica usando la fórmula del método de Euler.

```python
v_numerica = np.zeros(len(t))
v_numerica[0] = v0

dt = t[1] - t[0]  # Intervalo de tiempo

for i in range(1, len(t)):
    v_numerica[i] = v_numerica[i-1] + dv_dt(v_numerica[i-1], t[i-1]) * dt
```

# 3. Dibujo de la gráfica

```python
plt.plot(t, v_analitica, label='Solución Analítica')
plt.plot(t, v_numerica, 'o', label='Solución Numérica (Euler)', markersize=4)
plt.xlabel('Tiempo')
plt.ylabel('v(t)')
plt.legend()
plt.show()
```
